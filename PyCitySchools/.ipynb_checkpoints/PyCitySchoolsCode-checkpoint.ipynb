```python
# District Summary
# Perform the necessary calculations and then create a high-level snapshot of the district's key metrics in a DataFrame. 
#Include the following:

# Total number of unique schools, 
school_count = len(school_data)
print(f"Number of schools: {school_count}")

# print("Number of schools:",school_count) alternate code

# Total students, 
student_count = len(student_data)
print(f"Number of students: {student_count}")

# Total budget
total_budget = school_data ['budget'].sum()
print("Total Budget = ", total_budget)
# Average math score
average_mathscore = student_data["math_score"].mean()
print(average_mathscore)
# Average reading score 
average_readingscore = student_data["reading_score"].mean()
print(average_readingscore)
# % passing math (the percentage of students who passed math)
limit = 70
column_m = student_data["math_score"]
number_passing_math = column_m [column_m >= 70].count()
print('The number of students passing math was', number_passing_math)
print('The percentage of students passing math was', number_passing_math/student_count)
percentage_passing_math = number_passing_math/student_count * 100
# % passing reading (the percentage of students who passed readin
column_r = student_data["reading_score"]
number_passing_reading = column_r [column_r >= 70].count()

print('The number of students passing reading was', number_passing_reading)
print('The percentage of students passing reading was', number_passing_reading/student_count)
percentage_passing_reading = number_passing_reading/student_count * 100
# % overall passing (the percentage of students who passed math AND reading)
math_condition = (column_m >= 70)
reading_condition = (column_r >= 70)
number_passing_mathandreading = student_data [math_condition & reading_condition] ["Student ID"].count()
percentage_passing_mathandreading = number_passing_mathandreading/student_count * 100
# build new dataframe with these calculations
# df = pd.DataFrame(
# {"a" : [4, 5, 6],
# "b" : [7, 8, 9],
# "c" : [10, 11, 12]},
# index = [1, 2, 3])

district_summary = pd.DataFrame ({
   "Total Schools" : [school_count],
    "Total Students": [student_count],
    "Total Budget": [total_budget],
    "Average Math Score": [average_mathscore],
    "Average Reading Score": [average_readingscore],
    "% Passing Math": [percentage_passing_math],
    "% Passing Reading": [percentage_passing_reading],
    "% Overall Passing" : [percentage_passing_mathandreading] 
})

district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

district_summary
```
